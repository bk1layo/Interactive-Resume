```python
df = pd.read_csv("Life_Expectancy_preprocessed.csv")
df = df.iloc[:, 1:]

df
```

```python
df.describe()
```

```python
# Correlation analysis between the target variable(life expectancy), and the remaining variables.
# You only need to check the last row or last column.

# This is to figure out what variables affect the variable we are trying to predict (life expectancy).
# For example, this is to exclude unnecessary variables such as 'Year' from the model.
# When it comes to life expectancy, 'Year' is not a very convincing variable.

import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = df.iloc[:, 1:].corr()

colormap = plt.cm.PuBu
plt.figure(figsize=(12, 10))
plt.title("Pearson Correlation of Features", y = 1.05, size = 15)
sns.heatmap(correlation_matrix, linewidths = 0.1, vmax = 1.0,
           square = True, cmap = colormap, linecolor = "white", annot = True, annot_kws = {"size" : 8})

# The reason for df.iloc[ : , 1: ] is to exclude country names.
```

```python
# Extract values whose correlation exceeds threshold(We can change it arbitrarily, ex: 0.5)
# The closer the absolute value is to 1, the greater the correlation.

target_column = 'Life expectancy '
threshold = 0.4
columns_above_threshold = correlation_matrix.index[abs(correlation_matrix[target_column]) > threshold]

# print result
print("\nVariables whose correlation with the target variable exceeds {}:".format(threshold))
print(columns_above_threshold)
```

Regression

```python
# Since the list(columns_above_threshold) contains target value(Life Expectancy), we need to remove it from the list.
# Because its correlation with itself is 1, it is included in the list.

columns_above_threshold = list(columns_above_threshold)
columns_above_threshold.remove('Life expectancy ')
```

```python
# chosen variables only
# Divide variables into X(independent) and y(target, dependent)

X = df[columns_above_threshold]
y = df.iloc[:, -1]                 # target value
y = y.values.reshape(-1, 1)

# for checking
X.head()
```

```python
y
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


# Save model performance for different ratios
train_ratios = np.arange(0.1, 1.0, 0.1)  # ratios (from 0.1 to 0.9, 0.1 interval)
r2_train = []
r2_test = []
rmse_train = []
rmse_test = []

for train_ratio in train_ratios:
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_ratio, random_state=42)
    
    # train model
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Performance evaluation during training
    y_pred = model.predict(X_train)
    score = r2_score(y_train, y_pred)
    r2_train.append(score)

    rmse = np.sqrt(mean_squared_error(y_train, y_pred))
    rmse_train.append(rmse)     # save rmse in the list


    # evaluate model
    y_pred = model.predict(X_test)
    score = r2_score(y_test, y_pred)
    r2_test.append(score)      # save score in the list

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    rmse_test.append(rmse)     # save rmse in the list

```

```python
# Draw a graph
# Overfitting starts after 0.5

plt.figure(figsize=(10, 6))
plt.axis([0.05, 0.95, 0.795, 0.83])
plt.plot(train_ratios, r2_train, marker='o', linestyle='-')
plt.plot(train_ratios, r2_test, marker='o', linestyle='-')
plt.plot([0.5, 0.5], [0.795, 0.83], color = 'red', linewidth = 2)
plt.title('Performance comparison according to ratio change of training dataset')
plt.xlabel('Train Set Ratio')
plt.ylabel('R2 score')
plt.legend(["train", "test"], loc ="upper right")
plt.grid(True)
plt.show()
```

```python
# The RMSE of the test dataset increases slightly after 0.5. 

plt.figure(figsize=(10, 6))
plt.axis([0.05, 0.95, 3.4, 4.01])
plt.plot(train_ratios, rmse_train, marker='o', linestyle='-')
plt.plot(train_ratios, rmse_test, marker='o', linestyle='-')
plt.plot([0.5, 0.5], [3.4, 4.01], color = 'red', linewidth = 2)
plt.title('Loss comparison according to ratio change of training dataset')
plt.xlabel('Train Set Ratio')
plt.ylabel('RMSE')
plt.legend(["train", "test"], loc ="upper right")
plt.grid(True)
plt.show()
```

The model performs best at a test set ratio of 0.4 to 0.6.
I expected more than 0.9 but not as high as I expected.
And when it comes to error, it grdually increases slightly.

This is because underfitting occurs when there are few training sets, and overfitting occurs when there are too many training sets, making it difficult to predict test sets.

Generally, the proportion of training sets is often 0.6 or 0.7, but in this case 0.5 looks best when comparing performance and error.
